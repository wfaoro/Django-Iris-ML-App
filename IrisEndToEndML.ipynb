```python
from pandas.core.frame import DataFrame
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from urllib.request import urlretrieve
import pandas as pd

#Load iris dataset
iris = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
urlretrieve(iris)
iris = pd.read_csv(iris, sep=',', names=["sepal_length", "sepal_width", "petal_length", "petal_width", "class"])
iris
```

```python
#Split into training and testing data - 80/20 split respectively
X = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = iris["class"]

X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.20, random_state = 1)

#Lets fit SVC model on iris dataset
model = SVC(gamma = 'auto')
model.fit(X_train, Y_train)

# Making predictions on validation dataset
predictions = model.predict(X_test)

# Evaluating predictions
print(accuracy_score(Y_test, predictions))

# Pickle model
pd.to_pickle(model,r'C:\Users\William\MLapps\new_model.pickle')

# Unpickle model
model = pd.read_pickle(r'C:\Users\William\MLapps\new_model.pickle')
# Read a pickle pd.read_pickle('model.pkl')

# Take input from user
sepal_length = float(input("Enter sepal length: "))
sepal_width = float(input("Enter sepal width: "))
petal_length = float(input("Enter petal length: "))
petal_width = float(input("Enter petal width: "))

result = model.predict([[sepal_length, sepal_width, petal_length, petal_width]])
print(result)
```
